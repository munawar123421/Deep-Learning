```python
# =========================
# Step 0: Imports & Config
# =========================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import (
    accuracy_score, roc_auc_score, confusion_matrix, classification_report
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Config
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

DATA_PATH = "/kaggle/input/graduate-admissions/Admission_Predict_Ver1.1.csv"
THRESHOLD = 0.70        # stricter cutoff for Admit
DIM_REDUCTION = "pca"   # options: "pca", "lda", "none"
N_COMPONENTS = 5        # number of components for PCA
EPOCHS = 50
BATCH_SIZE = 32
N_SPLITS = 10

print("Config:")
print(f"  THRESHOLD={THRESHOLD}, DIM_REDUCTION='{DIM_REDUCTION}', "
      f"EPOCHS={EPOCHS}, BATCH_SIZE={BATCH_SIZE}, N_SPLITS={N_SPLITS}\n")

# ====================================
# Step 1: Load & Inspect the Dataset
# ====================================
df = pd.read_csv(DATA_PATH)
print("First 5 rows:")
print(df.head(), "\n")

print("Shape:", df.shape)
print("\nMissing values per column:")
print(df.isna().sum(), "\n")

# ===========================================
# Step 2: Feature/Target, Binarize the label
# ===========================================
X = df.drop(columns=["Serial No.", "Chance of Admit "], errors="ignore")
y = (df["Chance of Admit "] >= THRESHOLD).astype(int)

print("Feature columns:", list(X.columns), "\n")
print("Target distribution (0=Reject, 1=Admit):", dict(Counter(y)), "\n")

# =====================================
# Step 3: Build the ANN (3 hidden layers)
# =====================================
def build_model(input_dim):
    model = Sequential()
    model.add(Dense(10, activation='relu', input_dim=input_dim))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(5, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# =====================================
# Step 4: 10-fold Stratified CV + Train
# =====================================
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)

fold_histories = []
fold_metrics = []
trained_models = []

for fold, (tr_idx, va_idx) in enumerate(skf.split(X, y), start=1):
    print("="*60)
    print(f"Fold {fold}/{N_SPLITS}")

    # Split
    X_tr, X_va = X.iloc[tr_idx].copy(), X.iloc[va_idx].copy()
    y_tr, y_va = y.iloc[tr_idx].copy(), y.iloc[va_idx].copy()

    # Scale features
    scaler = StandardScaler()
    X_tr_sc = scaler.fit_transform(X_tr)
    X_va_sc = scaler.transform(X_va)

    # Apply feature reduction
    if DIM_REDUCTION == "pca":
        reducer = PCA(n_components=N_COMPONENTS, random_state=SEED)
        X_tr_red = reducer.fit_transform(X_tr_sc)
        X_va_red = reducer.transform(X_va_sc)

        if fold == 1:
            print("\nExplained variance by PCA components:")
            print(reducer.explained_variance_ratio_)
            print("Cumulative variance explained:", np.cumsum(reducer.explained_variance_ratio_))

    elif DIM_REDUCTION == "lda":
        reducer = LDA(n_components=1)  # 2-class problem → max 1 component
        X_tr_red = reducer.fit_transform(X_tr_sc, y_tr)
        X_va_red = reducer.transform(X_va_sc)

        if fold == 1:
            print("\nLDA class separation achieved.")

    else:
        X_tr_red, X_va_red = X_tr_sc, X_va_sc

    # Build model
    model = build_model(input_dim=X_tr_red.shape[1])

    if fold == 1:
        print("\nModel Summary (Fold 1):")
        model.summary(print_fn=lambda x: print("   " + x))

    # Train
    history = model.fit(
        X_tr_red, y_tr,
        validation_data=(X_va_red, y_va),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=0
    )

    fold_histories.append(history)
    trained_models.append(model)

    # Evaluate
    va_probs = model.predict(X_va_red, verbose=0).ravel()
    va_pred = (va_probs >= 0.5).astype(int)

    acc = accuracy_score(y_va, va_pred)
    try:
        auc = roc_auc_score(y_va, va_probs)
    except ValueError:
        auc = np.nan

    cm = confusion_matrix(y_va, va_pred)
    report = classification_report(y_va, va_pred, digits=4)

    fold_metrics.append({"fold": fold, "accuracy": acc, "roc_auc": auc})

    print(f"  Accuracy: {acc:.4f} | ROC-AUC: {auc:.4f}")
    print("  Confusion Matrix:\n", cm)
    print("  Classification Report:\n", report)

    # Plot training curves
    fig, ax = plt.subplots(1, 2, figsize=(12,4))
    ax[0].plot(history.history['accuracy'], label='train_acc')
    ax[0].plot(history.history['val_accuracy'], label='val_acc')
    ax[0].set_title(f'Fold {fold} - Accuracy')
    ax[0].legend()

    ax[1].plot(history.history['loss'], label='train_loss')
    ax[1].plot(history.history['val_loss'], label='val_loss')
    ax[1].set_title(f'Fold {fold} - Loss')
    ax[1].legend()
    plt.show()

# =====================================
# Step 5: Cross-fold Aggregation
# =====================================
metrics_df = pd.DataFrame(fold_metrics)
print("\nCross-fold results:")
print(metrics_df)
print("Mean Accuracy:", metrics_df['accuracy'].mean())
print("Mean ROC-AUC:", metrics_df['roc_auc'].mean())
```
